```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import f1_score, precision_score, recall_score
from ucimlrepo import fetch_ucirepo
```

```python
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17)
X = pd.DataFrame(breast_cancer_wisconsin_diagnostic.data.features) 
y = pd.DataFrame(breast_cancer_wisconsin_diagnostic.data.targets) 
y = y['Diagnosis'].map({'M': 1, 'B': 0})
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Create a model that uses only the first principal component of the data
pca = PCA(n_components=1)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
tModel = DecisionTreeClassifier(min_samples_leaf=2, min_samples_split=5, max_depth=2, random_state=42)
tModel.fit(X_train_pca, y_train)
y_pred = tModel.predict(X_test_pca)
```

```python
print("\n")
print("Using the first  principal component:")
print("F1 Score: {:.4f}".format(f1_score(y_test, y_pred)))
print("Precision: {:.4f}".format(precision_score(y_test, y_pred)))
print("Recall: {:.4f}".format(recall_score(y_test, y_pred)))
print("False Positives (FP): {}".format(np.sum((y_pred == 1) & (y_test == 0))))
print("True Positives (TP): {}".format(np.sum((y_pred == 1) & (y_test == 1))))
print("False Positive Rate (FPR): {:.4f}".format(np.sum((y_pred == 1) & (y_test == 0))/(np.sum((y_pred == 1) & (y_test == 0))+np.sum((y_pred == 0) & (y_test == 0)))))
print("True Positive Rate (TPR): {:.4f}".format(np.sum((y_pred == 1) & (y_test == 1))/(np.sum((y_pred == 1) & (y_test == 1))+np.sum((y_pred == 0) & (y_test == 1)))))
print("\n")
```

```python
#Create a model which uses the first and second principal components
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
tModel = DecisionTreeClassifier(min_samples_leaf=2, min_samples_split=5, max_depth=2, random_state=42)
tModel.fit(X_train_pca, y_train)
y_pred = tModel.predict(X_test_pca)
```

```python
print("\n")
print("Using first 2 principal components:")
print("F1 Score: {:.4f}".format(f1_score(y_test, y_pred)))
print("Precision: {:.4f}".format(precision_score(y_test, y_pred)))
print("Recall: {:.4f}".format(recall_score(y_test, y_pred)))
print("False Positives (FP): {}".format(np.sum((y_pred == 1) & (y_test == 0))))
print("True Positives (TP): {}".format(np.sum((y_pred == 1) & (y_test == 1))))
print("False Positive Rate (FPR): {:.4f}".format(np.sum((y_pred == 1) & (y_test == 0))/(np.sum((y_pred == 1) & (y_test == 0))+np.sum((y_pred == 0) & (y_test == 0)))))
print("True Positive Rate (TPR): {:.4f}".format(np.sum((y_pred == 1) & (y_test == 1))/(np.sum((y_pred == 1) & (y_test == 1))+np.sum((y_pred == 0) & (y_test == 1)))))
print("\n")
```
